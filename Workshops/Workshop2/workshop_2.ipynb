```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
images_book = pd.read_csv('BL-Flickr-Images-Book.csv')
```

```python
images_book.head()
```

```python
images_book_formatted = images_book[['Identifier', 'Place of Publication', 'Date of Publication',
                                     'Publisher', 'Title', 'Author', 'Flickr URL']]
images_book_formatted
```

```python
unwanted_characters_date = ['[', ']', '?', '-', '.', ',']

def clean_date(date):
    if isinstance(date, float):
        date = str(date)
    for char in unwanted_characters_date:
        date = date.replace(char, ' ')
    date = date.split(' ')[0]
    return date

images_book_formatted['Date of Publication'] = images_book_formatted['Date of Publication'].apply(clean_date)
images_book_formatted
```

```python
unwanted_characters_title = ['[', ']', '?', '-']

def clean_title(title):
    for char in unwanted_characters_title:
        title = title.replace(char, ' ')
    title = title.split('.')[0]
    return title

images_book_formatted['Title'] = images_book_formatted['Title'].apply(clean_title)
images_book_formatted
```
